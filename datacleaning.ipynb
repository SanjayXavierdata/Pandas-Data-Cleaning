```python
df=pd.read_excel("Customer Call List.xlsx")
df
```

```python
df.info()
```

Removing Duplicates

```python
df=df.drop_duplicates()
df
```

Removing Columns

```python
df=df.drop(columns="Not_Useful_Column")
df
```

Strip

```python
df["Last_Name"]=df["Last_Name"].str.strip("123/._")
df
```

Standardizing Data

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:x[0:3]+ '-'+x[3:6]+'-'+x[6:10])
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

Splitting Columns

```python
df[["street_address", "State", "zip_code"]] = (
    df["Address"].str.split(",", n=2, expand=True)
)
df
```

Standardizing

```python
df["Paying Customer"] = df["Paying Customer"].replace({
    "Y": "Yes",
    "N": "No"
})
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({
    "Y": "Yes",
    "N": "No"
})
df
```

Filtering Down Rows of Data

```python
df=df.fillna('')
df
```

```python
for x in df.index:
  if df.loc[x,"Do_Not_Contact"]=='Yes':
    df.drop(x,inplace=True)

df
```

```python
for x in df.index:
  if df.loc[x,"Phone_Number"]=='':
    df.drop(x,inplace=True)

df
```

```python
df=df.reset_index(drop=True)
df
```
